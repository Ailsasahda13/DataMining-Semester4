```python
chapter5 = pd.read_csv("DatasetJobsheet5.csv")

chapter5 = chapter5[[
    "Family",
    "Hobbies",
    "Social_Club",
    "Political",
    "Professional",
    "Religious",
    "Support_Group"
]]

chapter5.head()
```

```python
chapter5 = chapter5.replace("Yes", True)
chapter5 = chapter5.replace("No", False)
```

```python
frequent_ch5 = fpgrowth(chapter5, min_support=0.05, use_colnames=True)
frequent_ch5.head()
```

```python
rules_ch5 = association_rules(frequent_ch5, metric="confidence", min_threshold=0.5)
rules_ch5.sort_values(by="confidence", ascending=False)
```
